```python
import os
import numpy as np
import pandas as pd
```

```python
# 設定 data_path
dir_data = './data/'
```

```python
f_app = os.path.join(dir_data, 'application_train.csv')
print('Path of read in data: %s' % (f_app))
app_train = pd.read_csv(f_app)
```

### 如果沒有想法，可以先嘗試找出剛剛例子中提到的問題的答案
#### 資料的 row 數以及 column 數

```python
(r,c) = app_train.shape
print((r,c))
```

#### 列出所有欄位

```python
lst = app_train.columns.tolist() 
print(lst)
```

#### 還有各種數之不盡的資料操作，重點還是取決於實務中遇到的狀況和你想問的問題，在馬拉松中我們也會陸續提到更多例子

```python
#找出最大值及最小值位置
a = app_train['AMT_INCOME_TOTAL'].idxmax()
b = app_train['AMT_INCOME_TOTAL'].idxmin()
print(a)
print(b)
```

```python
#判定函數
app_train['AMT_CREDIT'] = app_train['AMT_CREDIT'].where(app_train['AMT_CREDIT']>500000, other=0)
app_train['AMT_CREDIT']
```

```python
#添加新的 columns
app_train['new_col'] = range(len(app_train))
app_train['new_col'] = np.repeat(np.nan,len(app_train))
app_train['random'] = np.random.rand(len(app_train))
app_train['index_as_col'] = app_train.index
app_train.head(2)
```

```python
#改變欄順序
app_train = app_train[['CODE_GENDER', 'TARGET', 'SK_ID_CURR','NAME_CONTRACT_TYPE']]
app_train
```
